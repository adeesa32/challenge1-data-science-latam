### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Facturacion1=sum(tienda.Precio)
Facturacion2=sum(tienda2.Precio)
Facturacion3=sum(tienda3.Precio)
Facturacion4=sum(tienda4.Precio)
print("Facturación");
print("Tienda 1: ",Facturacion1)
print("Tienda 2: ",Facturacion2)
print("Tienda 3: ",Facturacion3)
print("Tienda 4: ",Facturacion4)
```

```python
import matplotlib.pyplot as plt
colors=['#1f77b4','#ff7f0e','#2ca02c','#d62728']
VectorTiendas=['Tienda1','Tienda2','Tienda3','Tienda4']
VectorFacturacion=[Facturacion1,Facturacion2,Facturacion3,Facturacion4]
FacturacionMillones=[Facturacion1/1000000,Facturacion2/1000000,Facturacion3/1000000,Facturacion4/1000000]
plt.bar(VectorTiendas,FacturacionMillones,color=colors)
#esta parte coloca las etiquetas de números sobre las barras
for i, valor in enumerate(FacturacionMillones):
    plt.text(i, valor + 10, f"${valor:,.0f}", ha='center', va='bottom', fontsize=11)
# i: posición, +10 distancia de la barra
plt.ylim(1000,1200)
plt.ylabel('Facturacion en Millones')
plt.title('Facturación por Tienda')
```

```python
plt.pie(FacturacionMillones,labels=VectorTiendas,colors=colors,autopct='%1.1f%%')
plt.title('Facturación por Tienda en porcentaje')
plt.show()
```

## Categorías vendidas por tienda

```python
import builtins
zip = getattr(builtins, 'zip')
zipped=zip(df.iloc[:,1].value_counts().keys(),df.iloc[:,1].value_counts())
zipped2=zip(df2.iloc[:,1].value_counts().keys(),df2.iloc[:,1].value_counts())
zipped3=zip(df3.iloc[:,1].value_counts().keys(),df3.iloc[:,1].value_counts())
zipped4=zip(df4.iloc[:,1].value_counts().keys(),df4.iloc[:,1].value_counts())
valoresTienda1=list(zipped)
valoresTienda2=list(zipped2)
valoresTienda3=list(zipped3)
valoresTienda4=list(zipped4)
print(valoresTienda1)
```

Las categorías con más ventas en todas las tiendas son: Muebles, Electrónicos y Juguetes. Siguen los Electródomésticos y los Deportes. Las categorías con menor venta son los Artículos para el hogar y los libros en la mayoría de las tiendas.

```python
#VectorCategorias=['Muebles','Electrónicos','Jugetes','Electrodomésticos','Deportes y diversión','Instrumentos musicales','Libros','Artículos para el hogar']
VectorCategorias=df.iloc[:,1].value_counts().keys()
print(valoresTienda1)
print(valoresTienda2)
print(valoresTienda3)
print(valoresTienda4)
vectorTienda1=[0,0,0,0,0,0,0,0]
vectorTienda2=[0,0,0,0,0,0,0,0]
vectorTienda3=[0,0,0,0,0,0,0,0]
vectorTienda4=[0,0,0,0,0,0,0,0]
print(len(VectorCategorias))
for i in range(0, len(VectorCategorias)): #para cada categoría ya en orden print(i)
  for tupla in valoresTienda1: #por cada tupla en valores Tienda
    if VectorCategorias[i]==tupla[0]: # si la categoría es igual al primer elemento de la tupla
        vectorTienda1[i]=tupla[1] #asigna el número a la misma posición del vector print(vectorTienda1)
  for tupla in valoresTienda2: #por cada tupla en valores Tienda
    if VectorCategorias[i]==tupla[0]: # si la categoría es igual al primer elemento de la tupla
        vectorTienda2[i]=tupla[1]
  for tupla in valoresTienda3: #por cada tupla en valores Tienda
    if VectorCategorias[i]==tupla[0]: # si la categoría es igual al primer elemento de la tupla
        vectorTienda3[i]=tupla[1]
  for tupla in valoresTienda4: #por cada tupla en valores Tienda
    if VectorCategorias[i]==tupla[0]: # si la categoría es igual al primer elemento de la tupla
        vectorTienda4[i]=tupla[1]
print(vectorTienda1)
print(vectorTienda2)
print(vectorTienda3)
print(vectorTienda4)


```
